# 1.Reading data and dividing it in to Train and Test dataset

#Data downloaded from: https://www.kaggle.com/snap/amazon-fine-food-reviews/data

```python
#reading of dataframe using pandas by read_csv

#Library to data operations
import pandas as pd
#library for mumpy operations
import numpy as np
#For Plots
import seaborn as sns
import matplotlib.pyplot as plt


data=pd.read_csv("Reviews1.csv")
data
```

```python
#removing all the rows that contain Score =3
#Because 3 is neither negative nor positive rating!
data = data[data.Score != 3]
data
```

```python
#here we are going to set score =1,2 as negative ie 0
#and score4.5 as positive ie 1
data.loc[(data.Score == 1), 'Score'] = 0
data.loc[(data.Score == 2), 'Score'] = 0
data.loc[(data.Score == 4), 'Score'] = 1
data.loc[(data.Score == 5), 'Score'] = 1

data
```

```python
#To calculate no. of positive and negative reviews in whole dataset provided
data.Score.value_counts()
```

```python
#Sorting rows on the basis of Time stamp
data=data.sort_values(['Time'], ascending=[True])
data.head()
```

```python
#Dividing dataset to train and Test dataset
import math
totalsize=len(data)
 
train_size=int(math.ceil(0.7*totalsize))

#training dataset
train=data.head(train_size)
#test dataset
test=data.tail(totalsize-train_size)
train.Score.value_counts()
```

```python
#dropping columns that are not necessary for visulaization! Feature Selection
train.drop(['Id'], inplace=True, axis=1)
train.drop(['ProfileName'], inplace=True, axis=1)
train.drop(['HelpfulnessNumerator'], inplace=True, axis=1)
train.drop(['HelpfulnessDenominator'], inplace=True, axis=1)
train.drop(['Time'], inplace=True, axis=1)
```

```python
#Showing last 5 records
train.tail()
```

```python
# converting all values of text and summary to list
train_text= train.Text.tolist()
train_summary = train.Summary.tolist()
train_text
```

```python
#Converting to list
test_text=test.Text.tolist()
test_text
```

# 2.Preprocessing

Cleaning of Text

```python
#making all the sentences to lower case
train_text = [element.lower() for element in train_text]
train_text
```

```python
#removal of all the html tags 
import re
def striphtml(data):
    p = re.compile(r'<.*?>') # here we have made a regular expression using 
    #compile function to remove html tags
    return p.sub('', data)# here it will be replaced by ''
for i in range(len(train_text)): #for every element of a train_text list it will run 
    train_text[i]=striphtml(train_text[i])
train_text
```

```python
#removal of all the special characters if present in text even numbers using regular expression
for i in range(len(train_text)):
    for k in train_text[i].split("\n"): # here we first split words of every sentence by using split 
        #command and split whereever we see space
        train_text[i]=(re.sub(r"[^a-zA-Z]+", ' ', k))# this is going to remove everything from 
        #which is not a letter ie a-z A-z
train_text
```

```python
#here we use nltk library to remove stopwords from text
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
#all the stopwords get stored in stop_words variable
stop_words=set(stopwords.words("english"))

for i in range(len(train_text)): #for every sentence in a list train_text it will run
    #breaking of sentence in to tokens ie words
    train_text_words=word_tokenize(train_text[i])
    s=""
    #for every word this for loop runs
    for w in train_text_words:
        #then we are checking if any stop word is present or not
        if w not in stop_words:
            #if not present it will be appended to sentence
            s=s+" "+w
         #here we will get a sentence without stopwords   
        train_text[i]=s
train_text
```

```python
print(train_text[0])
```

```python
#STEMMING
#stemming helps to reduce the words to their stem
#Snowball Stemmer is used here one of the techniques to do stemming
from nltk.stem.snowball import SnowballStemmer
from nltk.tokenize import word_tokenize

for i in range(len(train_text)):
    train_text_filtered=word_tokenize(train_text[i])
    stemmer=SnowballStemmer("english")
    for p in range(len(train_text_filtered)):
        train_text_filtered[p]=(stemmer.stem(train_text_filtered[p]))
    s=""
    for w in train_text_filtered:
        s=s+" "+w
    train_text[i]=s
```
